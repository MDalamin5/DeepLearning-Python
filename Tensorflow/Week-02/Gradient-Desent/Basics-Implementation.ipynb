```python
from sklearn.datasets import make_regression
import numpy as np
```

```python
X, y = make_regression(n_samples=4, n_features=1, n_informative=1, n_targets=1, noise=80, random_state=13)
```

```python
import matplotlib.pyplot as plt
plt.scatter(X, y)
```

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
reg.fit(X, y)
```

```python
plt.scatter(X, y)
plt.plot(X, reg.predict(X), color='red')
```

### lets apply Gradient Descent assuming slope is constant m = 78.34
### and lets assume the starting value for intercept b = 0

```python
y_pred = ((78.35 * X) + 0).reshape(4)
```

```python
y_pred
```

```python
plt.scatter(X, y)
plt.plot(X, reg.predict(X), color='red', label='OLS')
plt.plot(X, y_pred, color = '#00a65a', label='b=0')
plt.legend()
plt.show()
```

```python
m = 78.35
b = 0
# print(X.ravel()) 
lossSlope = -2 * np.sum(y - m*X.ravel() - b)
lossSlope
```

```python
lr = 0.1

step_size = lossSlope*lr
step_size
```

```python
b = b - step_size
b
```

```python
y_pred1 = ((78.35 * X) + b).reshape(4)

plt.scatter(X, y)
plt.plot(X, reg.predict(X), color='red', label='OLS')
plt.plot(X, y_pred1, color='#00a65a', label='b = {}'.format(b))
plt.plot(X, y_pred, color = '#00a65a', label='b=0')
plt.legend()
plt.show()
```

```python
# iteration 2
lossSlope = -2 * np.sum(y - m*X.ravel() - b)
lossSlope
```

```python
step_size = lossSlope * lr
step_size
```

```python
b = b - step_size
b
```

```python
y_pred2 = ((78.35 * X) + b).reshape(4)
```

```python
plt.scatter(X, y)
plt.plot(X, reg.predict(X), color='red', label='OLS')
plt.plot(X, y_pred2, color='#00a65a', label='b = {}'.format(b))
plt.plot(X, y_pred1, color='#00a65a', label='b = {}'.format(b))
plt.plot(X, y_pred, color = '#00a65a', label='b=0')
plt.legend()
plt.show()
```

```python
# iteration 3

lossSlope = -2 * np.sum(y - m*X.ravel() - b)
lossSlope
```

```python
step_size = lossSlope*lr
step_size
```

```python
b = b - step_size
b
```

```python
b = -100
m = 78.53
lr = 0.1

epochs = 10

for i in range(epochs):
    lossSlope = -2 * np.sum(y - m*X.ravel() - b)
    b = b - (lr*lossSlope)
    y_pred = m * X + b
    plt.plot(X, y_pred)

plt.scatter(X, y)
```
