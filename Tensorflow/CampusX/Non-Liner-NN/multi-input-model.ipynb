```python
from keras.models import Model
from keras.layers import *
```

```python
inputA = Input(shape=(32,))
inputB = Input(shape=(128, ))
```

```python
rX1 = Dense(8, activation='relu')(inputA)
rX2 = Dense(4, activation='relu')(rX1)
```

```python
lX1 = Dense(64, activation='relu')(inputB)
lX2 = Dense(32, activation='relu')(lX1)
lx3 = Dense(4, activation='relu')(lX2)
```

```python
combined = concatenate([rX2, lx3])
```

```python
pxr = Dense(2, activation='relu')(combined)
output = Dense(1, activation='relu')(pxr)
```

```python
model = Model(inputs=[inputA, inputB], outputs=output)
```

```python
model.summary()
```

```python
from keras.utils import plot_model
plot_model(model)
```
