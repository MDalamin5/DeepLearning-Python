```python
import pandas as pd
import tensorflow as tf
from tensorflow import keras
from keras.preprocessing import image
from keras.preprocessing.image import ImageDataGenerator
import cv2
```

```python
# img = image.load_img('cat.jpg', target_size=(200, 200))
img = cv2.imread('cat.jpg')
```

```python
import matplotlib.pyplot as plt

plt.imshow(img)
```

```python
img = cv2.resize(img,(256, 256))
```

```python
img.shape
```

```python
plt.imshow(img)
```

```python
datagen = ImageDataGenerator(
        rotation_range=40,
        width_shift_range=0.2,
        height_shift_range=0.2,
        shear_range=0.2,
        zoom_range=0.2,
        horizontal_flip=True,
        fill_mode='constant'
)
```

```python
input_batch = img.reshape(1, 256, 256, 3)
```

```python
print(type(img))
```

```python
i=0
for output in datagen.flow(input_batch, batch_size=1, save_to_dir='augData'):
    i = i + 1
    if i == 10:
        break
```

```python
i=0
for output in datagen.flow(input_batch, batch_size=1, save_to_dir='augData1'):
    i = i + 1
    if i == 10:
        break
```
